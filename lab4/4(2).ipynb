```python
from collections import deque

class Node:
    """Klasa reprezentująca węzeł drzewa binarnego."""
    def __init__(self, data=None, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def bfs(root):
    if root is None:
        return
    
    # Utwórz pustą kolejkę i dodaj do niej korzeń
    queue = deque()
    queue.append(root)

    # Dopóki kolejka nie jest pusta
    while queue:
        # Pobierz pierwszy element z kolejki
        node = queue.popleft()
        # Wyświetl wartość aktualnego węzła
        print(node.data, end=" ")

        # Dodaj lewe i prawe dziecko do kolejki, jeśli istnieją
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

# Testowanie
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

print("BFS:", end=" ")
print
bfs(root)
```

```python
class Node:
    """Klasa reprezentująca węzeł drzewa binarnego."""
    def __init__(self, data=None, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def dfs(root):
    if root is None:
        return
    
    # Utwórz pusty stos i dodaj do niego korzeń
    stack = []
    stack.append(root)

    # Dopóki stos nie jest pusty
    while stack:
        # Pobierz ostatni element ze stosu
        node = stack.pop()
        # Wyświetl wartość aktualnego węzła
        print(node.data, end=" ")

        # Dodaj prawe i lewe dziecko do stosu, jeśli istnieją
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)

# Testowanie
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

print("DFS traversal:", end=" ")
dfs(root)
```

```python
class Node:
    """Klasa reprezentująca węzeł drzewa binarnego."""
    def __init__(self, data=None, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

    def __str__(self):
        return str(self.data)

def create_subtree_from_leaf(leaf):
    if leaf is None:
        return None
    
    # Tworzymy nowy węzeł dla aktualnego liścia
    new_node = Node(leaf.data)
    
    # Rekurencyjnie budujemy lewe i prawe poddrzewo
    new_node.left = create_subtree_from_leaf(leaf.left)
    new_node.right = create_subtree_from_leaf(leaf.right)
    
    return new_node

# Przykładowe wykorzystanie
root = None # puste drzewo
root = Node("start") # drzewo z jednym węzłem

# Ręczne budowanie większego drzewa.
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Wybieramy liść, np. liść o wartości 5
leaf_to_copy = root.left

# Tworzymy nowe drzewo binarne z wybranego liścia
new_tree = create_subtree_from_leaf(leaf_to_copy)

# Wyświetlamy nowe drzewo binarne
def display_tree(node, level=0):
    if node is not None:
        display_tree(node.right, level + 1)
        print("  " * level + str(node.data))
        display_tree(node.left, level + 1)

print("Nowe drzewo binarne:")
display_tree(new_tree)
```

```python
class Node:
    """Klasa reprezentująca węzeł drzewa binarnego."""
    def __init__(self, data=None, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

    def __str__(self):
        return str(self.data)

# Funkcja zliczająca liczbę węzłów na każdym poziomie i liczbę liści
def count_nodes_per_level(root):
    if root is None:
        return
    
    # Lista pomocnicza do BFS
    queue = [(root, 0)]
    
    # Inicjalizacja słownika do zliczania węzłów na każdym poziomie
    nodes_per_level = {}
    
    # Licznik liści
    leaf_count = 0
    
    while queue:
        node, level = queue.pop(0)
        
        # Zliczanie węzłów na danym poziomie
        if level in nodes_per_level:
            nodes_per_level[level] += 1
        else:
            nodes_per_level[level] = 1
        
        # Sprawdzenie czy węzeł jest liściem
        if node.left is None and node.right is None:
            leaf_count += 1
        
        # Dodanie dzieci do kolejki
        if node.left:
            queue.append((node.left, level + 1))
        if node.right:
            queue.append((node.right, level + 1))
    
    return nodes_per_level, leaf_count

# Przykładowe drzewo
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.left.left.left = Node(8)
root.left.left.right = Node(9)

# Liczenie węzłów na każdym poziomie i liści
nodes_per_level, leaf_count = count_nodes_per_level(root)

# Wyświetlanie wyników
print("Liczba węzłów na każdym poziomie:")
for level, count in nodes_per_level.items():
    print(f"Poziom {level}: {count} węzłów")

print(f"\nLiczba liści: {leaf_count}")
```

```python
class Node:
    """Klasa reprezentująca węzeł drzewa binarnego."""
    def __init__(self, data=None, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

    def __str__(self):
        return str(self.data)

# Funkcja wyznaczająca długość najkrótszej ścieżki od korzenia do liścia
# oraz wyświetlająca wszystkie liście na tej głębokości
def shortest_path_to_leaf(root):
    if root is None:
        return None, []
    
    # Lista pomocnicza do BFS
    queue = [(root, 1)]
    
    # Flagi do śledzenia pierwszego liścia na najkrótszej ścieżce
    shortest_path_found = False
    shortest_path_level = None
    
    # Lista do przechowywania liści na najkrótszej ścieżce
    shortest_path_leaves = []
    
    while queue:
        node, level = queue.pop(0)
        
        # Sprawdzamy czy znaleźliśmy pierwszy liść na najkrótszej ścieżce
        if not shortest_path_found and node.left is None and node.right is None:
            shortest_path_found = True
            shortest_path_level = level
        
        # Jeśli jesteśmy już na poziomie pierwszego liścia na najkrótszej ścieżce,
        # zapisujemy kolejne liście na tym poziomie
        if shortest_path_found and level == shortest_path_level:
            shortest_path_leaves.append(node.data)
        
        # Dodanie dzieci do kolejki
        if node.left:
            queue.append((node.left, level + 1))
        if node.right:
            queue.append((node.right, level + 1))
    
    return shortest_path_level, shortest_path_leaves

# Przykładowe drzewo
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.left.left.left = Node(8)
root.left.left.right = Node(9)

# Wyznaczanie najkrótszej ścieżki od korzenia do liścia i wyświetlanie liści na tej głębokości
shortest_path_length, shortest_path_leaves = shortest_path_to_leaf(root)

if shortest_path_length is not None:
    print(f"Długość najkrótszej ścieżki od korzenia do liścia: {shortest_path_length}")
    print("Liście na tej głębokości:")
    for leaf in shortest_path_leaves:
        print(leaf)
else:
    print("Drzewo jest puste.")
```

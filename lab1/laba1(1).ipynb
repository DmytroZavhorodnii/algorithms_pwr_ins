```python
import numpy as np
# Ustawienia rozmiaru macierzy
liczba_studentow = 5  # Możesz dostosować tę wartość
liczba_przedmiotow = 5  # Możesz dostosować tę wartość
macierz_ocen = np.random.uniform(low=2.0, high=5.5, size=(liczba_studentow , liczba_przedmiotow ))
macierz_ocen = np.round(macierz_ocen * 2) / 2


# Generowanie i wyświetlanie macierzy ocen
print("Macierz ocen:")
print(macierz_ocen)
```

```python
#1
prog = 3
n_przedmiotow = 2
nie_zdali =0

for i in range(liczba_studentow):
    # TRUE-FALSE
    bool = macierz_ocen[i]< prog
    #print(np.count_nonzero(bool))
    # False=0
    if np.count_nonzero(bool) >= n_przedmiotow:
        nie_zdali+=1

print("NIE ZDALI", nie_zdali)

```

```python
#2

# Найти индекс строки с самой большой средней

#матрица средних
#print(macierz_ocen.mean(axis=1))
index = np.argmax(macierz_ocen.mean(axis=1))
print(macierz_ocen[index])
```

```python
#3
najwyzsza_ocena = np.max(macierz_ocen)
print(najwyzsza_ocena)

max_sum = []
for k in range(liczba_studentow):
    count_per_row = np.sum(macierz_ocen[k] == najwyzsza_ocena)
    max_sum.append(count_per_row)

#print(np.array(max_sum))
index_ms = np.argmax(macierz_ocen.mean(axis=1))
print(macierz_ocen[index_ms])
```

```python
# Tworzenie histogramu 

import matplotlib.pyplot as plt

for przedmiot in range(liczba_przedmiotow):
    oceny_przedmiotu = macierz_ocen[:, przedmiot]
    
    # Oblicz histogram
    hist, edges = np.histogram(oceny_przedmiotu, bins=np.arange(2, 6.5, 0.5))

    # Wyświetl histogram
    plt.figure()
    plt.bar(edges[:-1], hist, width=0.5, align='center', edgecolor='black')
    plt.title(f'Histogram przedmiotu {przedmiot + 1}')
    plt.xlabel('Ocena')
    plt.ylabel('Liczba studentów')
    plt.show()
```

```python
def schodkowa_zredukowana(macierz):
    n, m = macierz.shape
    #print(macierz.shape)
    # Eliminacja Gaussa
    for k in range(min(n, m - 1)):
        # Szukanie maksymalnej wartości w kolumnie poniżej obecnej pozycji
        maks_wartosc_indeks = np.argmax(np.abs(macierz[k:, k])) + k

        # Zamiana wierszy, aby umieścić maksymalną wartość na przekątnej
        macierz[[k, maks_wartosc_indeks]] = macierz[[maks_wartosc_indeks, k]]

        # Normalizacja obecnego wiersza
        macierz[k] = macierz[k] / macierz[k, k]

        # Eliminacja pozostałych elementów w kolumnie
        for i in range(k + 1, n):
            macierz[i] -= macierz[i, k] * macierz[k]

    # Redukcja wsteczna
    for i in range(n - 1, 0, -1):
        for j in range(i - 1, -1, -1):
            macierz[j] -= macierz[j, i] * macierz[i]

    return macierz

# Przykładowa macierz
n = 3
macierz = np.random.rand(n, n + 1)
print("Macierz przed redukcją:")
print(macierz)

# Sprowadzenie do postaci schodkowej zredukowanej
macierz_schodkowa_zredukowana = schodkowa_zredukowana(macierz.copy())
print("\nMacierz po redukcji:")
print(macierz_schodkowa_zredukowana)
```

```python
import numpy as np

# Przykładowe dane dla pierwszej macierzy (zakupy)
macierz_zakupy = np.array([
    [1, 101, 2],  # klient 1, towar 101, 2 sztuki
    [1, 102, 1.5],  # klient 1, towar 102, 1.5 kg
    [2, 101, 1],  # klient 2, towar 101, 1 sztuka
    [2, 103, 3]  # klient 2, towar 103, 3 sztuki
])

# Przykładowe dane dla drugiej macierzy (opisy towarów)
macierz_opisy_towarow = np.array([
    [101, 10.0, 'sztuki'],  # towar 101, cena 10.0 za sztukę, sprzedaż na sztuki
    [102, 5.0, 'kg'],  # towar 102, cena 5.0 za kg, sprzedaż na wagę
    [103, 2.5, 'sztuki']  # towar 103, cena 2.5 za sztukę, sprzedaż na sztuki
])

# Sprawdzenie, czy numery towarów istnieją w macierzy towarów
numery_towarow = macierz_opisy_towarow[:, 0]

for numer_towaru in macierz_zakupy[:, 1]:
    if numer_towaru not in numery_towarow:
        print(f"Błąd: Towar o numerze {numer_towaru} nie istnieje w macierzy opisów towarów.")

# Sprawdzenie, czy towary sprzedawane na sztuki są liczbami całkowitymi
for wiersz in macierz_zakupy:
    numer_towaru = wiersz[1]
    ilosc = wiersz[2]
    index_towaru = np.where(macierz_opisy_towarow[:, 0] == numer_towaru)[0][0]
    jednostka_sprzedazy = macierz_opisy_towarow[index_towaru, 2]
    
    if jednostka_sprzedazy == 'sztuki' and not ilosc.is_integer():
        print(f"Błąd: Towar {numer_towaru} sprzedawany na sztuki, ale ilość {ilosc} nie jest liczbą całkowitą.")
```

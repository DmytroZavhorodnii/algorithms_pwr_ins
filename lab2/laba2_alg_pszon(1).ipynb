```python
#bubl sort
import random
import time

L = [random.randint(10, 1000) for _ in range(random.randint(1, 100))]
print("Original List:", L)

times = []  # список для хранения времени выполнения каждой итерации

for i in range(len(L)):
    start_time = time.time()  # начало измерения времени
    for j in range(0, (len(L) - 1) - i):
        if L[j] > L[j + 1]:
            temp = L[j]
            L[j] = L[j + 1]
            L[j + 1] = temp
            # print(L)  # вывод списка после каждого шага сортировки
    end_time = time.time()  # конец измерения времени
    iteration_time = end_time - start_time
    times.append(iteration_time)  # добавляем время итерации в список

# вывод среднего и максимального времени
average_time = sum(times) / len(times)
max_time = max(times)

print("Average time per iteration:", average_time)
print("Maximum time per iteration:", max_time)
```

```python
import matplotlib.pyplot as plt

def bubble_sort(L):
    times = []
    for i in range(len(L)):
        start_time = time.time()
        for j in range(0, len(L) - 1 - i):
            if L[j] > L[j + 1]:
                L[j], L[j + 1] = L[j + 1], L[j]
        end_time = time.time()
        iteration_time = end_time - start_time
        times.append(iteration_time)
    return times

def insertion_sort(L):
    times = []
    for i in range(len(L)):
        start_time = time.time()
        for j in range(i):
            if L[i] < L[j]:
                L[i], L[j] = L[j], L[i]
        end_time = time.time()
        iteration_time = end_time - start_time
        times.append(iteration_time)
    return times

def selection_sort(L):
    times = []
    for i in range(len(L)):
        start_time = time.time()
        for j in range(i + 1, len(L)):
            if L[i] > L[j]:
                L[i], L[j] = L[j], L[i]
        end_time = time.time()
        iteration_time = end_time - start_time
        times.append(iteration_time)
    return times

# Długości ciągów
lengths = [10, 20, 50, 100, 200, 500, 1000]

# Średnie i maksymalne czasy dla każdego algorytmu
bubble_average_times = []
bubble_max_times = []
insertion_average_times = []
insertion_max_times = []
selection_average_times = []
selection_max_times = []

for length in lengths:
    L = [random.randint(1000, 10000000) for _ in range(length)]
    bubble_times = bubble_sort(L.copy())
    bubble_average_times.append(sum(bubble_times) / len(bubble_times))
    bubble_max_times.append(max(bubble_times))

    L = [random.randint(1000, 10000000) for _ in range(length)]
    insertion_times = insertion_sort(L.copy())
    insertion_average_times.append(sum(insertion_times) / len(insertion_times))
    insertion_max_times.append(max(insertion_times))

    L = [random.randint(1000,10000000) for _ in range(length)]
    selection_times = selection_sort(L.copy())
    selection_average_times.append(sum(selection_times) / len(selection_times))
    selection_max_times.append(max(selection_times))

# Tworzenie wykresu

plt.figure(figsize=(10, 6))
plt.plot(lengths, bubble_average_times, label='Bubble Sort (Average)')
plt.plot(lengths, bubble_max_times, label='Bubble Sort (Max)')
plt.plot(lengths, insertion_average_times, label='Insertion Sort (Average)')
plt.plot(lengths, insertion_max_times, label='Insertion Sort (Max)')
plt.plot(lengths, selection_average_times, label='Selection Sort (Average)')
plt.plot(lengths, selection_max_times, label='Selection Sort (Max)')
plt.xlabel('Length of List')
plt.ylabel('Time (seconds)')
plt.title('Comparison of Sorting Algorithms')
plt.legend()
plt.grid(True)
plt.xticks(lengths, labels=lengths)
plt.tight_layout()
plt.show()
```

```python
import random
import time

# Funkcja implementująca standardowy algorytm sortowania bąbelkowego
def bubble_sort_standard(L):
    times = []  # Lista do przechowywania czasu wykonania każdej iteracji
    for i in range(len(L)):
        start_time = time.time()  # Początek pomiaru czasu
        for j in range(0, len(L) - 1 - i):
            if L[j] > L[j + 1]:
                L[j], L[j + 1] = L[j + 1], L[j]
        end_time = time.time()  # Koniec pomiaru czasu
        iteration_time = end_time - start_time
        times.append(iteration_time)  # Dodanie czasu iteracji do listy
    return times

# Funkcja implementująca algorytm sortowania bąbelkowego z przerywaniem, gdy nie nastąpiła żadna zamiana w bieżącym przejściu
def bubble_sort_early_stop(L):
    times = []  # Lista do przechowywania czasu wykonania każdej iteracji
    for i in range(len(L)):
        start_time = time.time()  # Początek pomiaru czasu
        swapped = False  # Flaga wskazująca, czy doszło do zamiany w bieżącej iteracji
        for j in range(0, len(L) - 1 - i):
            if L[j] > L[j + 1]:
                L[j], L[j + 1] = L[j + 1], L[j]
                swapped = True
        end_time = time.time()  # Koniec pomiaru czasu
        iteration_time = end_time - start_time
        times.append(iteration_time)  # Dodanie czasu iteracji do listy
        if not swapped:  # Jeśli nie doszło do żadnej zamiany w bieżącej iteracji, przerywamy sortowanie
            break
    return times

# Funkcja implementująca algorytm sortowania bąbelkowego, w którym w każdym przejściu po ciągu porównania naiwnie realizujemy do końca ciągu
def bubble_sort_naive_comparison(L):
    times = []  # Lista do przechowywania czasu wykonania każdej iteracji
    for i in range(len(L)):
        start_time = time.time()  # Początek pomiaru czasu
        for j in range(0, len(L) - 1):  # Naiwne wykonanie porównań do końca ciągu w każdym przejściu
            if L[j] > L[j + 1]:
                L[j], L[j + 1] = L[j + 1], L[j]
        end_time = time.time()  # Koniec pomiaru czasu
        iteration_time = end_time - start_time
        times.append(iteration_time)  # Dodanie czasu iteracji do listy
    return times

# Funkcja generująca listę losowych liczb
def generate_random_list(length):
    return [random.randint(10, 1000) for _ in range(length)]

# Długości list do przetestowania
lengths = [10, 20, 50, 100, 200, 500, 1000]

# Wyniki dla standardowego algorytmu sortowania bąbelkowego
standard_results = []

for length in lengths:
    L = generate_random_list(length)
    times = bubble_sort_standard(L.copy())
    standard_results.append(sum(times) / len(times))  # Średni czas dla każdej długości listy

# Wyniki dla algorytmu sortowania bąbelkowego z przerywaniem
early_stop_results = []

for length in lengths:
    L = generate_random_list(length)
    times = bubble_sort_early_stop(L.copy())
    early_stop_results.append(sum(times) / len(times))  # Średni czas dla każdej długości listy

# Wyniki dla algorytmu sortowania bąbelkowego z naiwnymi porównaniami
naive_comparison_results = []

for length in lengths:
    L = generate_random_list(length)
    times = bubble_sort_naive_comparison(L.copy())
    naive_comparison_results.append(sum(times) / len(times))  # Średni czas dla każdej długości listy

# Wyświetlenie wyników
print("Standard Bubble Sort Results:", standard_results)
print("Bubble Sort with Early Stop Results:", early_stop_results)
print("Bubble Sort with Naive Comparisons Results:", naive_comparison_results)



# Plot the results
plt.plot(lengths, standard_results, label='Standard Bubble Sort')
plt.plot(lengths, early_stop_results, label='Bubble Sort with Early Stop')
plt.plot(lengths, naive_comparison_results, label='Bubble Sort with Naive Comparisons')

# Add labels and title
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.title('Comparison of Bubble Sort Variants')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
#ostatnie----------------------------------
import random
import time
import math

def bubble_sort(L):
    start_time = time.time()
    for i in range(len(L)):
        for j in range(0, len(L) - 1 - i):
            if L[j] > L[j + 1]:
                L[j], L[j + 1] = L[j + 1], L[j]
    end_time = time.time()
    return end_time - start_time

def insertion_sort(L):
    start_time = time.time()
    for i in range(len(L)):
        key = L[i]
        j = i - 1
        while j >= 0 and L[j] > key:
            L[j + 1] = L[j]
            j -= 1
        L[j + 1] = key
    end_time = time.time()
    return end_time - start_time

def selection_sort(L):
    start_time = time.time()
    for i in range(len(L)):
        min_index = i
        for j in range(i + 1, len(L)):
            if L[j] < L[min_index]:
                min_index = j
        L[i], L[min_index] = L[min_index], L[i]
    end_time = time.time()
    return end_time - start_time

def run_performance_tests(algorithm, n_values, repetitions=10):
    results = []
    for n in n_values:
        total_time = 0
        for _ in range(repetitions):
            arr = [random.randint(1, 1000000) for _ in range(n)]
            total_time += algorithm(arr)
        average_time = total_time / repetitions
        theoretical_time = 0 if average_time == 0 else (n * math.log(n)) / average_time
        results.append((n, average_time, theoretical_time))
    return results

# Długości ciągów
n_values = [10, 100, 1000]

# Wykonanie testów wydajnościowych dla każdego algorytmu
bubble_results = run_performance_tests(bubble_sort, n_values)
insertion_results = run_performance_tests(insertion_sort, n_values)
selection_results = run_performance_tests(selection_sort, n_values)

# Wyświetlenie wyników
print("Bubble Sort Results:")
for result in bubble_results:
    print("n:", result[0], "| t(n):", result[1], "| (n·log(n))/t(n):", result[2])

print("\nInsertion Sort Results:")
for result in insertion_results:
    print("n:", result[0], "| t(n):", result[1], "| (n·log(n))/t(n):", result[2])

print("\nSelection Sort Results:")
for result in selection_results:
    print("n:", result[0], "| t(n):", result[1], "| (n·log(n))/t(n):", result[2])
```

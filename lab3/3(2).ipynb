```python
import random
import time

class Node:
    def __init__(self, typ, zlozonosc):
        self.typ = typ
        self.zlozonosc = zlozonosc
        self.next = None

class Kolejka:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, typ, zlozonosc):
        new_node = Node(typ, zlozonosc)
        if self.rear is None:
            self.front = new_node
            self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.front is None:
            return None
        removed_node = self.front
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return removed_node

    def is_empty(self):
        return self.front is None

class Okno:
    def __init__(self, numer, typ, czas_zajecia):
        self.numer = numer
        self.typ = typ
        self.czas_zajecia = czas_zajecia
        self.liczba_obsluzonych = 0

    def obsluz_zadanie(self, zadanie):
        self.czas_zajecia += zadanie.zlozonosc
        self.liczba_obsluzonych += 1

def testuj_urzad(okienka, kolejka):
    start_time = time.time()

    liczba_iteracji = 0
    while any(okno.czas_zajecia > 0 for okno in okienka) or not kolejka.is_empty():
        for okno in okienka:
            if okno.czas_zajecia == 0:
                current = kolejka.front
                while current:
                    if current.typ == okno.typ:
                        zadanie = kolejka.dequeue()
                        okno.obsluz_zadanie(zadanie)
                        break
                    current = current.next
            elif okno.czas_zajecia > 0:
                okno.czas_zajecia -= 1
        liczba_iteracji += 1

    end_time = time.time()
    return liczba_iteracji, end_time - start_time

def generuj_kolejke(liczba_zadan, porzadek):
    kolejka = Kolejka()
    if porzadek == 'rosnacy':
        for _ in range(liczba_zadan):
            typ_zadania = random.choice(['A', 'B', 'C'])
            zlozonosc_zadania = random.randint(1, 12)
            kolejka.enqueue(typ_zadania, zlozonosc_zadania)
    elif porzadek == 'malejacy':
        for _ in range(liczba_zadan, 0, -1):
            typ_zadania = random.choice(['A', 'B', 'C'])
            zlozonosc_zadania = random.randint(1, 12)
            kolejka.enqueue(typ_zadania, zlozonosc_zadania)

    return kolejka

def testuj_dla_porzadku(liczba_zadan, porzadek):
    kolejka = generuj_kolejke(liczba_zadan, porzadek)
    
    iteracje_urzad_1, czas_urzad_1 = testuj_urzad(urzad_1, kolejka)

    return czas_urzad_1

# Tworzenie okienek dla trzech różnych wersji urzędów
urzad_1 = [
    Okno(1, 'A', 5),
    Okno(2, 'A', 5),
    Okno(3, 'A', 5),
    Okno(4, 'B', 5),
    Okno(5, 'B', 5),
    Okno(6, 'B', 5),
    Okno(7, 'C', 5),
    Okno(8, 'C', 5),
    Okno(9, 'C', 5),
    Okno(10, 'E', 5)
]


# Testowanie dla kolejki w porządku rosnącym
czas_urzad_1_rosnacy= testuj_dla_porzadku(3000, 'rosnacy')

# Testowanie dla kolejki w porządku malejącym
czas_urzad_1_malejacy= testuj_dla_porzadku(3000, 'malejacy')

# Wyświetlanie wyników
print("Wersja urzędu 1:")
print("Czas obsługi kolejki dla kolejności rosnącej:", czas_urzad_1_rosnacy, "sekund")
print("Czas obsługi kolejki dla kolejności malejącej:", czas_urzad_1_malejacy, "sekund")
print()
```
